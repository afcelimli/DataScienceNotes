```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'], kde=False, bins=30)
```

```python
sns.jointplot('total_bill', 'tip', tips, kind='reg')
```

```python
sns.pairplot(tips, hue='sex')
```

```python
sns.kdeplot(tips['tip'])
```

```python
sns.rugplot(tips['tip'])
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(x='day',y='tip',data=tips,hue='sex')
```

```python
sns.violinplot(x='day',y='tip',data=tips,hue='sex',split=True)
```

```python
sns.stripplot(x='day',y='tip',data=tips,hue='sex')
```

```python
sns.swarmplot(x='day',y='tip',data=tips)
```

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
tc = tips.corr()
tc
```

```python
sns.heatmap(tc, annot=True)
```

```python
fp = flights.pivot_table(values='passengers',index='month',columns='year')
fp
```

```python
sns.heatmap(fp)
```

```python
sns.clustermap(fp)
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(sns.barplot)
g.map_lower(sns.kdeplot)
```

```python
g = sns.FacetGrid(tips, row='time', col='smoker')
g.map(sns.distplot, 'total_bill')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex')
```

```python
#plt.figure(figsize=(12,8))
sns.set_context(context='poster')
sns.set_style(style='whitegrid')
sns.barplot(x='day',y='tip',data=tips,palette='coolwarm')
sns.despine()
```
