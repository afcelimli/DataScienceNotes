# K Nearest Neighbors Project 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
```

# DATA

```python
df = pd.read_csv('Datasets/KNN_Project_Data')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.pairplot(df, hue='TARGET CLASS', diag_kind=None)
```

# Standardize the Features

```python
scaler = StandardScaler()
scaler.fit(df.drop('TARGET CLASS', axis=1), df['TARGET CLASS'])
scaled_features = scaler.transform(df.drop('TARGET CLASS', axis=1))
```

```python
df_scaled = pd.DataFrame(scaled_features, columns=df.columns[:-1])
df_scaled.head()
```

# Model

```python
X_train, X_test, y_train, y_test = train_test_split(df_scaled, df['TARGET CLASS'], test_size=0.3, random_state=101)
```

## Model with 1 neighbor

```python
knn_model = KNeighborsClassifier(n_neighbors=1)
```

```python
knn_model.fit(X_train, y_train)
```

```python
predictions = knn_model.predict(X_test)
```

```python
print(confusion_matrix(y_test, predictions))
```

```python
print(classification_report(y_test, predictions))
```

## Neighbor Selection

```python
error_rate = []

for i in range(1,101):
    tmp_model = KNeighborsClassifier(n_neighbors=i)
    tmp_model.fit(X_train, y_train)
    predictions = tmp_model.predict(X_test)
    error = np.mean(y_test != predictions)
    error_rate.append(error)
```

```python
plt.figure(figsize=(12,6))
plt.plot(range(1,101), error_rate)
```

```python
final_model = KNeighborsClassifier(n_neighbors=33)
final_model.fit(X_train, y_train)
predictions = final_model.predict(X_test)
```

```python
print(confusion_matrix(y_test, predictions))
```

```python
print(classification_report(y_test, predictions))
```
