```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
print(x)
```

```python
print(y)
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes1.plot(x,y)
axes1.set_title("First figure")
axes1.set_xlabel("X")
axes1.set_ylabel("Y")

axes2 = fig.add_axes([.2, .4, .3,.3])
axes2.plot(y,x, 'r--')
axes2.set_title("Second figure")
axes2.set_xlabel("X")
axes2.set_ylabel("Y")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12,6), dpi=50)
axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
fig.savefig("asd2.png")
```

```python
fig = plt.figure(figsize=(6,6))
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x**2,label="Squared")
axes.plot(x,x**3,label="Cubed")
axes.set_title("Exercise")
axes.set_xlabel("X")
axes.set_ylabel("Y")
axes.legend(loc=0)
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x, x**2, color="red", linestyle="-", linewidth=1, marker="o", alpha=.8)
```
